## Computation of Lennar-Jones Parameters

$\frac{\epsilon}{k_B} = \frac{T_C}{1.32}$

$\sigma = \biggl(\frac{0.129 \epsilon}{P_C}\biggr)^{(1/3)}$

$T_C = 593,75 K$

$P_C = 41,6 atm$

```python
Tc = 593.75 # K
Pc = 41.6 # atm
kb = 0.001985875	# kcal/(K * mol)
N = 1000
Na = 6.022*10**(23) 
rho = 8.7 # kg/cum
PM = 92.14

eps = kb*Tc/1.32     #units real ha le energie in kcal/mol
#sigma = (0.129*eps/Pc)**(1/3)    #units real ha le distanze in angstrom
# quindi : kcal/atm = 41.2928695 L = 0.0412928695 cum, fratto avogadro per togliere le moli e 
# moltiplicato per 10^30 per ottenere armstrom cubi
sigma = (0.129*eps/Pc*0.0412928695*10**30/Na)**(1/3) 

V = (N/Na*PM/1000)/rho*10**(30)
L = V**(1/3)

print('eps :', eps)
print('sigma :', sigma)
print('V :', V)
print('L :', L)
```

# Data Analysis

## TOLUENE 550

###Equilibration fluctuation

```python
import numpy as np
import matplotlib.pyplot as plt

# Caricamento del file
THERMO_file = np.genfromtxt('/content/thermodynamics_equilibration_550.dat', delimiter=' ')

time = THERMO_file[:, 0]
T = THERMO_file[:, 4]
P = THERMO_file[:, 5]
V = THERMO_file[:, 6]

# Calcolo della media
P_media = np.mean(P)

# Plot della pressione e del suo valore medio
plt.figure(1)
plt.plot(time, P)
plt.axhline(P_media)
print('P_media: ', P_media)

# Calcolo della media
T_media = np.mean(T)

# Plot della pressione e del suo valore medio
plt.figure(2)
plt.plot(time, T)
plt.axhline(T_media)
print('T_media: ', T_media)
```

### Diffusion coefficient

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

MSD_file = np.genfromtxt('/content/msd_550.dat', delimiter=' ')
time = MSD_file[:, 0]
msd = MSD_file[:, 1]
plt.figure(1)
plt.xlabel("t [fs]")
plt.ylabel("MSD [Angstrom^2]")
plt.title("MSD vs Tempo (Scala lineare)")
plt.plot(time,msd)

plt.figure(2)
plt.loglog(time,msd)
plt.xlabel("t [fs]")
plt.ylabel("MSD [Angstrom^2]")
plt.title("MSD vs Tempo (Scala logaritmica)")

def linea_retta(t,D):
  return 6*D*t

D, err = curve_fit(linea_retta,time,msd)
plt.figure(3)
plt.plot(time,msd)
plt.plot(time,linea_retta(time,D), '--')
plt.xlabel("t [fs]")
plt.ylabel("MSD [Angstrom^2]")
plt.title("MSD vs Tempo (Scala logaritmica)")
plt.show()
print(D)
```

### Ideal gas law  

```python
import numpy as np

# Caricamento del file
THERMO_file = np.genfromtxt('/content/thermodynamics_550.dat', delimiter=' ')

# Vettori colonna
T = THERMO_file[:, 4]
P = THERMO_file[:, 5]
V = THERMO_file[:, 6]

# Valori medi
T_sim = T.mean()
P_sim = P.mean()
V_sim = V.mean()*10**(-27) #conversione in litri

# Legge dei gas ideali
P_ideal = (1000/(6.022*10**23))*0.0821*T_sim/V_sim

# Scostamento percentuale
dP = (P_sim-P_ideal)/P_ideal*100

print('P_ideal: ', P_ideal,'atm') 
print('P_sim: ', P_sim,'atm') 
print('Scostamento percentuale: ', dP, '%')
```

# Diffusion Fit

```python
# Fit della diffusivit contro T
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

T = np.array([550, 573, 590, 620, 630])
D = np.array([0.18621044, 0.19406967, 0.19887452, 0.20861725, 0.20599098])

def func(x,a,b):
  return a*x+b

parametri,err = curve_fit(func,T,D)
print(parametri)

plt.plot(T,D,'o')
T = np.linspace(550,630,10000)
plt.plot(T,func(T,parametri[0],parametri[1]))
plt.show()

T = np.array([550, 573, 590, 620, 630])
y_values = func(T,parametri[0],parametri[1])
correlation_matrix = np.corrcoef(D, y_values)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2

print(r_squared)
```
